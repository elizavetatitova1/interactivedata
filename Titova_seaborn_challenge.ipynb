## Seaborn challenge for group 1

```python
titanic = pd.read_csv('titanic_clean.csv')
```

```python
titanic.head()
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare'
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5, hue = 'sex'
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5, hue = 'sex',
    col = 'survived'
)
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5, hue = 'sex',
    col = 'survived',
    palette = {'male': "blue", 'female': "red"}
)
```

```python
titanic['Sex'] = titanic['sex'].map({'male': 'Male', 'female': 'Female'})
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5, hue = 'Sex',
    col = 'survived',
    palette = {'Male': "blue", 'Female': "magenta"},
    size = 'pclass'
)
```

```python
titanic['survived_ind'] = titanic['survived'].map({0: 'did not survive', 1: 'survived'})
```

```python
sns.relplot(
    data = titanic,
    x = 'age', y = 'fare',
    alpha = 0.5, hue = 'survived_ind',
    col = 'survived',
    palette = {'did not survive': "black", 'survived': "green"},
    
)
```

## Seaborn challenge for group 2

```python
import matplotlib.pyplot as plt
```

```python
smh = pd.read_csv('Student Mental health.csv')
```

```python
smh.head()
```

```python
smh.groupby('Choose your gender')[['Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']].count()
```

```python
columns_to_plot = smh.columns[-4:-1]
category_order = ['Yes', 'No']
```

```python
for column in columns_to_plot:
    counts  = smh[column].value_counts()
    counts = counts.reindex(category_order, fill_value=0)
    print(counts)
    print()
    
```

```python
sns.countplot(smh, x = 'Do you have Depression?', stat = 'percent', hue = 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x = 'Do you have Anxiety?', stat = 'percent', hue = 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x = 'Do you have Panic attack?', stat = 'percent', hue = 'Did you seek any specialist for a treatment?')
```

```python
sns.countplot(smh, x = 'Do you have Depression?', stat = 'percent', hue = 'Choose your gender', palette = 'PuBuGn')
```
